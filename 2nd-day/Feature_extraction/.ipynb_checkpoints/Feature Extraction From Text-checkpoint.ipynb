```python
# The inverse document frequency (IDF) is a measure of how rare or common a word is in a corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = ['The dog ate a sandwich and I ate a sandwich','The wizard transfigured a sandwich']
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
```

```python
print (vectorizer.fit_transform(corpus).todense())
```

```python
# {u'sandwich': 2, u'wizard': 4, u'dog': 1, u'transfigured': 3, u'ate':
#    0}
```

```python
vectorizer.get_feature_names()
```

```python
vectorizer.get_stop_words()
```
